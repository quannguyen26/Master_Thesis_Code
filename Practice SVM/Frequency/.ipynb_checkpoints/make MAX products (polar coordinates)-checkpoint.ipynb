```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap,BoundaryNorm
import matplotlib.cm as cm
import pyart
import numpy as np
import warnings
import numpy.ma as ma
warnings.filterwarnings('ignore')
filename = 'PHA210802220004.RAWKLA7'
```

```python
# create the plot using RadarDisplay (recommended method)
radar = pyart.io.read_sigmet(filename)
Z=radar.fields['reflectivity']['data']
```

```python
mask_noises_Z=ma.masked_less(radar.fields['reflectivity']['data'],5)
mask_noises_CC=ma.masked_less(radar.fields['cross_correlation_ratio']['data'],0.85)
Zh_=np.ma.masked_array(Z,mask_noises_CC.mask)
Zh=np.ma.masked_array(Zh_,mask_noises_Z.mask)
```

```python
Zh[0:360,:].max()
```

```python
maxs=[]
for ray in range(0,360,1):
    r=[]
    for i in (ray,ray+360,ray+720,ray+1080,ray+1440):
        r.append(Zh[i,:])
    array=np.array(r)
    maxs.append(np.amax(array,axis=0))
Zh_max=np.array(maxs)
```

```python
Zh_max
```

```python
Zh_max.min()
```

```python
cmap=ListedColormap(['deepskyblue', 'dodgerblue','blue', 'palegreen','lime','limegreen','yellow','orange','darkorange','red','mediumvioletred','darkviolet','black'])
bounds=[-50,10,15,20,25,30,35,40,45,55,60,65,70,100]
norm = BoundaryNorm(bounds, cmap.N)
ticks  = [10,15,20,25,30,35,40,45,55,60,65,70]
```

```python
x = radar.range['data']*np.sin(np.deg2rad(radar.azimuth['data']))[:360,None]
y = radar.range['data']*np.cos(np.deg2rad(radar.azimuth['data']))[:360,None]
```

```python
fig=plt.figure(figsize=(5,4),dpi=300)
ax1=fig.add_subplot(1,1,1)
z2=ax1.pcolormesh(x,y,Zh[361:720,:],cmap=cmap,norm=norm)
ax1.set_ylim([-200000, 200000])
ax1.set_xlim([-200000, 200000])
plt.colorbar(z2,ax=ax1,ticks=ticks,extend='both',label="DBZ");
```
