```python
import pyart
RADAR_NAME = 'PHA210207124005.RAWSB71'
# read in the data
radar = pyart.io.read_sigmet(RADAR_NAME)
```

```python
dir(radar)
```

```python
radar=radar.extract_sweeps([1])
```

```python
radar.nsweeps
```

```python
import matplotlib.pyplot as plt
display = pyart.graph.RadarDisplay(radar)
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
display.plot('reflectivity', vmin=-32, vmax=64.)
#display.plot_range_rings([10, 20, 30, 40])
display.plot_cross_hair(5.)
plt.show()
```

```python
lat_0 = radar.latitude['data'][0]
lon_0 = radar.longitude['data'][0]
shape = (1, 601, 601)
grid = pyart.map.grid_from_radars(
            radar,
            grid_shape=shape, #Number of points in the grid (z, y, x)
            grid_limits=((2000.0, 12000.0), (-300000, 300000), (-300000, 300000)), # min-max tuong duong z,y,x
            grid_origin = (lat_0,lon_0),
            gridding_algo="map_gates_to_grid",
            weighting_function='BARNES2',
            fields=['reflectivity'])
```

```python
grid.fields['reflectivity']['data'].shape
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(grid.fields['reflectivity']['data'][0], origin='lower')
plt.show()
```
