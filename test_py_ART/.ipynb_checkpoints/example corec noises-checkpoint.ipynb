```python
import matplotlib.pyplot as plt
import pyart
import warnings
import numpy as np
warnings.filterwarnings('ignore')
RADAR_NAME = 'PHA210505201004.RAWC0KS'

# read in the data
radar = pyart.io.read_sigmet(RADAR_NAME)
```

```python
#gatefilter = pyart.filters.GateFilter(radar)
#gatefilter.exclude_below('reflectivity', 0)
```

```python
radar=radar.extract_sweeps([0])
```

```python
a=radar.fields['reflectivity']['data'][:,-2:]
```

Tìm mode của a (Zf)

```python
from scipy import stats
from scipy.stats import mstats
```

```python
Zf=mstats.mode(a,axis=None)[0][0]
```

Tìm Zs

```python
range_bins=radar.range['data']
```

```python
Zs=pyart.retrieve.compute_noisedBZ(360, noisedBZ_val=Zf, _range=range_bins, ref_dist=300., noise_field='noise_dbz')
```

```python
Zsplus4=Zs['data']+4
```

```python
import numpy.ma as ma
s=ma.masked_less(radar.fields['reflectivity']['data'],Zsplus4)
```

```python
radar.add_field_like('reflectivity', 'reflectivity_cor_noise',s,replace_existing=True)
```
