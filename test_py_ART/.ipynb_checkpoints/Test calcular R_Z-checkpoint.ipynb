```python
radar = radar.extract_sweeps([0])
```

```python
spec_at, cor_z = pyart.correct.attenuation.calculate_attenuation(
    radar,0,refl_field='reflectivity')
#spec_at['data'][np.where(spec_at['data'] < 0.0)] = 0.0
```

```python
est_R_z=pyart.retrieve.est_rain_rate_z(radar,alpha=0.0365, beta=0.625,refl_field='corrected_reflectivity_horizontal')
```

```python
from wradlib.trafo import idecibel,r_to_depth
```

```python
est_R_z['data']
```

```python
R_mm=r_to_depth(est_R_z['data'],600)
```

```python
R_mm
```

```python
est_R_z['data']=R_mm
est_R_z['units']='mm'
```

```python
est_R_z
```

```python
#radar.fields.pop('est_R_z_mm')
radar.add_field('est_R_z_mm',est_R_z)
```

```python
display = pyart.graph.RadarDisplay(radar)
fig = plt.figure(figsize = [15,10])
display.plot_ppi('est_R_z_mm')
```

```python
lat_0=radar.latitude['data']
lon_0=radar.longitude['data']
grid_shape=(1,301,301)
grid=pyart.map.grid_from_radars(
            radar,
            grid_shape=grid_shape, #Number of points in the grid (z, y, x) (12,301,301)
            grid_limits=((0, 12000), (-300000, 300000), (-300000, 300000)), # min-max tuong duong z,y,x
            grid_origin = (lat_0,lon_0),
            fields=['est_R_z_mm'])
```

```python
for i in range(grid_shape[1]):
    geog = pyart.core.cartesian_to_geographic_aeqd(grid.x["data"][i],grid.y["data"][i],lon_0,lat_0, R=6370997.0)
    grid.x["data"][i] = geog[0]
    grid.y["data"][i] = geog[1]
```

```python
R=grid.fields['est_R_z_mm']['data'][0]
```

```python
R.shape
```

```python
fig = plt.figure(figsize=(12,10))
ax = plt.subplot(111, aspect="equal")
pm = plt.pcolormesh(grid.x['data'], grid.y['data'],R)
plt.colorbar(pm, shrink=0.75)
```
